### Part B: Compare and Evaluate AI Techniques

In marketing, companies use Artificial Intelligence (AI) to better understand customers, predict what they might do next, and create personalised experiences. For the task of customer segmentation and behaviour prediction, three useful AI techniques are: **K-Means Clustering**, **Logistic Regression**, and **Artificial Neural Networks**. Each technique works differently and can be applied to different marketing problems. This section compares the strengths, weaknesses, inputs, and outputs of these three techniques and explains why K-Means Clustering will be used for the prototype in Part C.

---

### 1. K-Means Clustering (Unsupervised Learning)

**Goal:** K-Means is used to group customers into different segments based on similar behaviour or characteristics. It can use features like spending amount, number of visits, income level, or preferred products.

**How it helps in marketing:** K-Means can identify groups such as **high-value buyers**, **discount shoppers**, **loyal customers**, or **occasional visitors**. This allows companies to understand their audience and plan targeted marketing campaigns for each segment.

**Advantages:**
* Very simple and fast to use, even on large datasets (Kavlakoglu and Winland, 2024).
* Easy to understand because each group has a clear centre (centroid).
* Good for customer segmentation, where the goal is to find natural patterns in data.

**Disadvantages:**
* Sensitive to outliers, which can pull clusters in the wrong direction (Kavlakoglu and Winland, 2024).
* Requires choosing the number of clusters (k) before running the model.
* Works best with numerical data.
* Struggles when clusters are very different in size or density.

**Data Needed and Expected Output:** K-Means needs numerical data such as spending amounts, visit frequency, income, or age. The output is a set of clusters, each representing a type of customer. These clusters are usually shown using charts or scatter plots.

---

### 2. Logistic Regression (Supervised Learning)

**Goal:** Logistic Regression predicts the probability of a binary outcome, such as whether a customer will buy a product, click an ad, or stop using a service.

**How it helps in marketing:** This model can estimate how likely a customer is to respond to a campaign or how at risk they are of churning. Marketers can then focus their efforts on the right customers using data from behaviour, demographics, or campaign history (resonio, 2024).

**Advantages:**
* Very interpretable: marketers can easily understand which features influence the result.
* Robust and less sensitive to outliers.
* Works well with medium or large datasets and is easy to implement.

**Disadvantages:**
* Requires a large and good-quality dataset to work well (resonio, 2024).
* Assumes a linear relationship, meaning it struggles with complex patterns.
* Only suitable for binary or simple multi-category predictions.

**Data Needed and Expected Output:** Requires labelled data such as clicked/not clicked or churned/not churned. Outputs include a probability (e.g. customer has a 0.82 chance of churning) and a final YES/NO prediction.

---

### 3. Artificial Neural Networks (Supervised Deep Learning)

**Goal:** Artificial Neural Networks (ANNs) aim to learn complex relationships in customer data. They can predict many types of marketing outcomes and find deeper patterns than simple models.

**How it helps in marketing:** Businesses use ANNs for tasks such as predicting customer lifetime value, recommending products, understanding purchase behaviour, and analysing customer sentiment in reviews (Lee, F. 2021). Companies like Netflix and Spotify use neural networks to recommend content.

**Advantages:**
* Very good at identifying complex and nonlinear patterns.
* Works well with large and messy datasets.
* Can achieve high accuracy when enough training data is available.

**Disadvantages:**
* Hard to interpret; marketers cannot easily see how the model makes decisions ("Black Box" problem).
* Needs a lot of data to perform well.
* Can overfit if not trained properly (Domaleski, 2025).

**Data Needed and Expected Output:** Needs large amounts of customer data such as browsing history, transactions, and demographic information. Outputs include predictions, probability scores, and personalised recommendations.

---

### Conclusion

For this project, customer segmentation, **K-Means Clustering** is the most suitable because it is simple, computational efficient and interpretable, and very effective at grouping customers based on behaviour. While Neural Networks offer higher accuracy, they were rejected for this prototype due to the requirement for a massive, labelled dataset, which is not available. Logistic Regression was also discarded as the goal is segmentation (unsupervised exploration) rather than prediction (supervised classification).

#### B. Input Data & Pre-processing

**1. Data Source and Description**
The input data for this prototype is the **E-commerce Customer Behaviour Dataset** (`ecommerce_customer_data_custom_ratios`), sourced from open-source repositories from Kaggle (Jagtap, n.d.). This dataset contains transactional records from an online retailer.

* **Raw Format:** The data is provided in a **CSV (Comma Separated Values)** file.
* **Structure:** It is a **Time-Series Transaction Log**, where each row represents a single product purchase event. Key columns include `Customer ID`, `Purchase Date`, `Product Price`, `Quantity`, and `Returns`.
* **Volume:** The dataset contains approximately **250,000 records**, ensuring sufficient density for the machine learning algorithm to detect meaningful patterns.

**Key Features:**
* **Customer ID:** Unique identifier for each user.
* **Total Purchase Amount:** The monetary value of a single transaction.
* **Purchase Date:** Timestamp of the purchase.
* **Returns:** Indicator of whether an item was returned.
* **Churn:** Label indicating if a customer stopped buying.

**2. Data Pre-processing Pipeline**
Raw transactional data cannot be directly fed into the K-Means algorithm because the algorithm requires a single row per data point (customer). Therefore, extensive pre-processing was required to transform transactions into a "Customer Matrix."

**Step 1: Fixing Data Integrity (Math & Dates)**
* **Observation:** The `Total Purchase Amount` column in the raw file did not match the product of `Price * Quantity`.
* **Action Taken:** We recalculated the column using the formula: `Total Purchase Amount = Product Price * Quantity`.
* **Justification:** Relying on the original broken values would have severely skewed the "Monetary" feature in the clustering model, leading to false customer segments (e.g., labelling a low-value customer as a "VIP" due to inflated totals). Correcting this ensures the segmentation is based on actual spending behaviour.

**Step 2: Checking Missing IDs**
* **Observation:** Upon inspecting the dataset, it was essential to ensure `Customer ID` integrity.
* **Action Taken:** We removed all rows where the `Customer ID` was missing using `df.dropna(subset=['Customer ID'])`.
* **Justification:** This step is critical for **Customer Segmentation**. The K-Means algorithm requires us to aggregate data to the "Customer Level" (grouping by ID to calculate total spend and frequency). Without a unique ID, it is impossible to track a users history or build a behavioural profile (RFM score). Including anonymous transactions would create a distorted "Unknown User" cluster that provides no actionable marketing insight.

# K-MEANS

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(r"EcommerceData.csv")
```

```python
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'], format='mixed', dayfirst=True)
```

```python
df.head()
```

```python
X = pd.get_dummies(df)
```

```python
X
```

```python
# 1. LOAD AND CONVERT DATES (Run this first!)


# 2. Setup Snapshot Date
snapshot_date = X['Purchase Date'].max()

# --- Step A: Recency ---
# Find last purchase date
last_purchase_date = X.groupby('Customer ID')['Purchase Date'].max()

# Calculate days (This line will now work because dates are fixed)
recency = (snapshot_date - last_purchase_date).dt.days

# --- Step B: Frequency ---
frequency = X.groupby('Customer ID')['Customer ID'].count()

# --- Step C: Monetary ---
monetary = X.groupby('Customer ID')['Total Purchase Amount'].sum()

# --- Step D: Combine ---
rfm = pd.DataFrame({
    'Recency': recency,
    'Frequency': frequency,
    'Monetary': monetary
})

print(rfm.head())
```

```python
rfm
```

```python
# 2. Scale the small RFM data (This will succeed)
scaler = StandardScaler()
scaled_rfm = scaler.fit_transform(rfm)
```

```python
scaled_rfm
```

```python
#Determine Optimal K (Elbow Method Plot)
inertia = []
range_k = range(1, 11)
for k in range_k:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_rfm)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range_k, inertia, 'bx-')
plt.xlabel('k (Number of Clusters)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True);

# --- 4. Apply Final K-Means (K=4) ---
optimal_k = 4
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(scaled_rfm)

# Assign clusters back to the RFM dataframe
rfm['Cluster'] = cluster_labels
#plt.savefig('elbowMethod.png')
```

### Part D: Software Testing and Evaluation

#### 1. Testing and Verification

K-Means is an unsupervised learning technique; there are no pre-existing "correct" labels (e.g., True/False) to test against, unlike supervised learning where the algorithm will find a $Y$ label to test against $X$. Instead, testing focused on validating the **optimal structure** of the model and verifying the **semantic logic** of the results. The following testing methods were applied:

**Test A: Determining Optimal Clusters (The Elbow Method)**
In K-Means clustering, identifying the optimal number of clusters is not straightforward. Therefore, I used the **Elbow Method** to plot the **Within-Cluster Sum of Squares (WCSS)** (Inertia) against increasing $k$ values, looking for a point where the improvement slows down significantly.

* **Observation:** The resulting plot showed a sharp decrease in inertia which began to flatten out significantly at **$k=4$**.
* **Conclusion:** This plot confirmed that 4 is the mathematically optimal number of clusters for this specific dataset. Choosing fewer clusters would under-fit the data (merging distinct groups), while choosing more would over-fit (splitting natural groups arbitrarily).



**Test B: Visual Inspection (Scatter Plot Verification)**
To test how the clustered data set would look, I plotted a 2D Scatter Plot of **Recency vs. Monetary**, colouring each data point by its assigned cluster label.

* **Observation:** The visualization showed clear, non-overlapping regions for the high-value and low-value segments.
* **Verification:** This visual confirmation proves that the mathematical distance calculations in the K-Means algorithm successfully partitioned the dataset in the vector space. The "VIP" customers (High Money, Low Recency) were clearly separated from the "Lost" customers (Low Money, High Recency).



---

#### 2. Evaluation and Insights

**Assessing Effectiveness**
The K-Means implementation successfully solved the stated problem of customer segmentation by partitioning the raw transactional data into four distinct behavioural groups. The model effectively translated complex, multi-dimensional data (Recency, Frequency, Monetary) into a simplified "Customer Scorecard." This achieves the project goal of enabling targeted marketing strategies rather than a "one-size-fits-all" approach.

**Interpretation of Results (The 4 Personas)**
By analysing the cluster centroids (mean values), we can interpret the results in the context of E-commerce marketing. The algorithm identified four clear customer personas:

* **The "VIP Champions" (High Monetary, Low Recency):**
    * **Insight:** These are the most valuable assets to the business. They trust the brand and purchase regularly.
    * **Actionable Strategy:** Focus on **Retention**. Offer them a premium loyalty program, early access to new products, or "refer-a-friend" incentives. Do *not* waste discount coupons on them, as they are already willing to pay full price.

* **The "Loyal Casuals" (Medium Monetary, Low Recency):**
    * **Insight:** These customers are active but spend less per visit than VIPs. They have the potential to be "upsold."
    * **Actionable Strategy:** Focus on **Increasing Basket Size**. Use recommendation engines (e.g., "Customers who bought X also bought Y") or bundle deals to nudge them toward higher spending.

* **The "At-Risk" Customers (High Spending, High Recency):**
    * **Insight:** These were once good customers who are drifting away (churning).
    * **Actionable Strategy:** Focus on **Re-engagement**. Send a targeted "We Miss You" email campaign with a time-limited discount.

* **The "Lost / Low Value" (Low Monetary, High Recency):**
    * **Insight:** These users likely made a single purchase long ago and never returned.
    * **Actionable Strategy:** **Deprioritize**. Marketing budget should not be concentrated here. Automated, low-cost emails can be sent, and expensive ad retargeting should be disabled for this segment.

---

#### 3. Strengths and Limitations

**Strengths of the Implementation**
* **Interpretability:** The RFM model combined with K-Means produced highly logical clusters. It is easy to explain to stakeholders *why* a customer is in a specific group (e.g., "Because they haven't shopped in 200 days").
* **Computational Efficiency:** As noted in the literature review, K-Means is computationally fast. The prototype processed the 250,000+ records in just a few seconds, demonstrating it is suitable for real-time marketing applications.
* **Scalability:** The code structure allows new transaction data to be added easily. The pipeline is automated and repeatable.

**Limitations and Future Improvements**
* **Static Snapshot Bias:** The current model calculates Recency based on a fixed "Snapshot Date". This means the model degrades over time. **Improvement:** In a production environment, this script would need to run daily or weekly to update customer scores dynamically.
* **Constraint of Euclidean Distance:** K-Means assumes clusters are spherical and of similar size. If the "VIP" group is very small and dense while the "Low Value" group is massive and spread out, K-Means might struggle to separate them perfectly.
* **Lack of "Product" Context:** The model only looks at *how much* was spent, not *what* was bought. A customer buying baby diapers has different needs than one buying video games, even if their RFM scores are identical.

### Part E: Equity, Diversity, Inclusion (EDI) and Sustainability

#### 1. Critical Evaluation of EDI: Bias, Fairness, and Inclusion

While the K-Means clustering algorithm is mathematically objective, its implementation in a marketing context carries significant ethical risks regarding **Bias and Fairness**. The model segments customers purely on financial metrics (Recency, Frequency, Monetary). This creates an inherent "economic bias," therefore customers from lower socioeconomic backgrounds may be consistently categorized as "Low Value" or "At Risk" simply due to limited disposable income. Those groups could be just normal demographics that make purchases on basis of needs. If marketing strategies solely reward the high-spending "VIP" cluster with discounts and exclusive benefits, the system inadvertently reinforces advantages for high-income customers, penalizing lower-income groups for their financial status rather than their loyalty.

Furthermore, regarding **Accessibility and Inclusion**, the model relies exclusively on digital transaction logs. This creates a data blind spot known as the **"Digital Divide"** (OECD, 2019). Moreover, demographics who uses cash as primary mode of purchase or in-store purchases often including the elderly or unbanked populations are invisible to the algorithm. Consequently, strategic decisions based on this model will disproportionately favour digital natives while discarding vulnerable groups who cannot access e-commerce sites.

#### 2. Environmental Sustainability and Resource Analysis

From a **Computational Resource** perspective, the chosen K-Means technique is highly efficient. As noted in Part B, its computational complexity is low compared to deep learning models, allowing the prototype to process the dataset with minimal energy consumption. This makes it a sustainable choice for frequent re-training without incurring a heavy carbon footprint.

However, the primary environmental impact lies in **Data Storage**. E-commerce transaction logs grow exponentially, and storing years of historical data in cloud data centres demands significant continuous electricity for servers and cooling systems (Monserrate, 2022). To reduce this environmental impact in a production deployment, it is suggested to implement a strict **Data Lifecycle Policy**. Since RFM analysis typically relies on recent behaviour, data older than 24 months offers diminishing returns. Archiving or deleting aged data would significantly reduce active storage requirements, thereby lowering the overall energy cost of the system.
